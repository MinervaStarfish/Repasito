# 2. Medidas de Tendencia Central

En el contexto del Análisis Exploratorio de Datos (EDA), las medidas de tendencia central son estadísticas descriptivas que proporcionan un indicador representativo de la posición central o típica de un conjunto de datos. Las medidas más comunes de tendencia central incluyen la media, la mediana y la moda.

* La media aritmética es la suma de todos los valores dividida por el número de observaciones, y es especialmente sensible a valores extremos.

* La mediana es el valor que divide a los datos en dos mitades, donde el 50% de las observaciones están por debajo y el 50% están por encima, lo que la hace más robusta ante valores atípicos.

* La moda representa el valor más frecuente en un conjunto de datos.

Estas medidas ofrecen una visión general de la ubicación central de los datos, lo que ayuda a comprender la distribución y la forma de la información durante la exploración y análisis inicial de un conjunto de datos.

inspeccionemos el atributo de `price_usd` **(variable numérica continua)** de los autos listados en el dataset:

```python
df['price_usd'].mean()
```

```python
df['price_usd'].median()
```

```python
df['price_usd'].plot.hist(bins=200)
```

El parámetro 'bins' en la función de histograma `df['price_usd'].plot.hist(bins=20)` especifica la cantidad de contenedores (o intervalos) en los que se dividirá el rango de valores de la variable 'price_usd' para construir el histograma.

En otras palabras, 'bins' controla la granularidad con la que se visualizan los datos en el histograma. Si 'bins' se establece en un número mayor, se obtendrán más contenedores, lo que dará una representación más detallada de la distribución. Por el contrario, un número menor de 'bins' agrupará los datos en menos intervalos, proporcionando una vista más general de la distribución.

En el ejemplo `df['price_usd'].plot.hist(bins=20)`, se ha elegido dividir el rango de valores de 'price_usd' en 20 contenedores. Este valor puede ajustarse según las características específicas de tus datos y tus preferencias visuales. Experimentar con diferentes valores de 'bins' te permitirá obtener una representación visual más informativa de la distribución de la variable.

# 3. Medidas de Dispersión

En el contexto del Análisis Exploratorio de Datos (EDA), las medidas de dispersión son estadísticas descriptivas que proporcionan información sobre la variabilidad o dispersión de un conjunto de datos. Estas medidas cuantifican la extensión o dispersión de los valores alrededor de una medida central, como la media o la mediana. Las medidas de dispersión comunes incluyen la varianza y la desviación estándar, que representan la dispersión promedio de los datos respecto a la media. Una dispersión más alta indica una mayor variabilidad entre los valores, mientras que una dispersión más baja sugiere que los datos tienden a estar más agrupados alrededor de la medida central.

```python
# Desviación estandar
df['price_usd'].std()
```

```python
# Rango = valor max - valor min
rango = df['price_usd'].max() - df['price_usd'].min()
rango
```

```python
# Cuartiles
median = df['price_usd'].median()
Q1 = df['price_usd'].quantile(q=0.25)
Q3 = df['price_usd'].quantile(q=0.75)
min_val = df['price_usd'].quantile(q=0)
max_val = df['price_usd'].quantile(q=1.0)
print(min_val, Q1, median, Q3, max_val)
```

**Relación entre Histograma y Boxplot:**

Ambas gráficas proporcionan información sobre la dispersión de los datos, pero de manera diferente.

El histograma muestra la forma general de la distribución, mientras que el boxplot destaca medidas resumen clave y facilita la identificación de valores atípicos.

Al colocar un boxplot debajo de un histograma, se logra una representación más completa y eficiente de la distribución de los datos.

```python
import matplotlib.pyplot as plt

# Configuración del estilo de Seaborn
sns.set(style="whitegrid")

# Crear una figura con dos subgráficos
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6), sharex=True, gridspec_kw={"height_ratios": (0.7, 0.3)})

# Histograma en el primer subgráfico
sns.histplot(df['price_usd'], bins=100, ax=axes[0])
axes[0].set_title('Histograma')

# Boxplot en el segundo subgráfico
sns.boxplot(df['price_usd'], orient='h', ax=axes[1])
axes[1].set_title('Boxplot')

# Ajustar el diseño para que las gráficas estén más cerca
plt.tight_layout()

# Mostrar la visualización
plt.show()
```

# 4. Procesamiento de Datos Categóricos

**4.1 Concepto de Variable Ordinal:**

Una variable ordinal es aquella que tiene categorías que siguen un orden natural. Por ejemplo, niveles de educación (primaria, secundaria, universitaria) o grados de satisfacción (insatisfecho, neutral, satisfecho).

**4.1 Concepto de Variable Nominal:**

Una variable nominal es un tipo de variable categórica donde las categorías no tienen un orden natural o jerárquico entre ellas. Cada categoría es simplemente diferente de las otras, sin implicar ningún tipo de ranking o secuencia.
Características clave de las variables nominales:

* No tienen orden inherente.
* Las categorías son mutuamente excluyentes.
* Solo se pueden contar o categorizar, no ordenar.


Veamos un ejemplo para ilustrar esto:

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

# Datos de ejemplo: niveles de satisfacción
data = np.array(['bajo', 'medio', 'alto', 'medio', 'bajo'])

print("Datos originales:", data)
```

```python
# Usando LabelEncoder
le = LabelEncoder()
le_encoded = le.fit_transform(data)

print("LabelEncoder:", le_encoded)
```

```python
# Usando OrdinalEncoder
oe = OrdinalEncoder(categories=[['bajo', 'medio', 'alto']])
oe_encoded = oe.fit_transform(data.reshape(-1, 1))

print("OrdinalEncoder:", oe_encoded)
```

**Observaciones:**

LabelEncoder asigna los valores basándose en el orden alfabético: 'alto' = 0, 'bajo' = 1, 'medio' = 2.
OrdinalEncoder respeta el orden que especificamos: 'bajo' = 0, 'medio' = 1, 'alto' = 2.

La diferencia clave aquí es que `OrdinalEncoder` nos permite especificar el orden natural de las categorías, lo cual es crucial para variables ordinales. `LabelEncoder`, por otro lado, siempre usará el orden alfabético, lo cual puede no ser apropiado para variables ordinales.

En situaciones donde el orden es importante (como en niveles de satisfacción o grados educativos), `OrdinalEncoder` es más adecuado porque preserva la relación ordinal entre las categorías. `LabelEncoder` es más apropiado para variables categóricas nominales donde no hay un orden inherente (como colores o nombres de países).

### Detalles adicionales acerca de OrdinalEncoder:

1. `OrdinalEncoder` proporciona un parámetro `encoded_missing_value` para codificar los valores faltantes sin necesidad de de preprocesamiento adicional o manejo manual de los valores nulos. Esta característica permite una codificación más directa y eficiente de los valores faltantes dentro del propio proceso de transformación del `OrdinalEncoder`. Al especificar este parámetro, se puede asignar automáticamente un valor predeterminado a todos los valores faltantes en los datos, simplificando así el flujo de trabajo y reduciendo la necesidad de pasos adicionales en el preprocesamiento de los datos.

```python
# Datos con valores faltantes
data = np.array(['bajo', 'medio', np.nan, 'medio', 'bajo', 'alto', np.nan], dtype=object).reshape(-1, 1)

# Configurar OrdinalEncoder para manejar valores faltantes
oe = OrdinalEncoder(encoded_missing_value=-1)

# Ajustar y transformar los datos
encoded = oe.fit_transform(data)

print("Datos originales:", data.ravel())
print("Datos codificados:", encoded.ravel())
print("Categorías:", oe.categories_)
```

2. El parámetro `handle_unknown='use_encoded_value'` en OrdinalEncoder proporciona una solución elegante para manejar categorías desconocidas durante la transformación de datos, sin necesidad de reajustar el codificador o realizar un preprocesamiento adicional. Esta opción permite al codificador asignar un valor predefinido a cualquier categoría que no haya sido vista durante la fase de entrenamiento, en lugar de generar un error. Cuando se utiliza junto con el parámetro unknown_value, permite especificar exactamente qué valor numérico se debe asignar a estas categorías nuevas o desconocidas.

```python
# Datos de entrenamiento
train_data = np.array(['bajo', 'medio', 'alto', 'medio', 'bajo']).reshape(-1, 1)

# Configurar OrdinalEncoder para manejar categorías desconocidas
oe = OrdinalEncoder(categories=[['bajo', 'medio', 'alto']], handle_unknown='use_encoded_value', unknown_value=-2)

# Ajustar con los datos de entrenamiento
oe.fit(train_data)

# Datos de prueba con una nueva categoría
test_data = np.array(['bajo', 'medio', 'muy alto', 'bajo', 'extremadamente alto']).reshape(-1, 1)

# Transformar los datos de prueba
encoded = oe.transform(test_data)

print("Datos de entrenamiento:", train_data.ravel())
print("Datos de prueba:", test_data.ravel())
print("Datos codificados:", encoded.ravel())
print("Categorías:", oe.categories_)
```

**4.3 Codificación One-Hot**

Otra posibilidad para convertir características categóricas a características que se pueden usar con los estimadores de scikit-learn es usar la codificación "one-hot" o "dummy". Este tipo de codificación se puede obtener con `OneHotEncoder`, que transforma cada característica categórica con n valores posibles en n características binarias, con una de ellas siendo 1 y todas las demás 0.


```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
X = [['hombre', 'de EE. UU.', 'usa Safari'], ['mujer', 'de Europa', 'usa Firefox']]
enc.fit(X)

enc.transform([['mujer', 'de EE. UU.', 'usa Safari'], ['hombre', 'de Europa', 'usa Safari']]).toarray()
```

* Por defecto, los valores que puede tomar cada característica se infieren automáticamente a partir del conjunto de datos y se pueden encontrar en el atributo `categories_`.

* `OneHotEncoder` también admite características categóricas con valores faltantes al considerar los valores faltantes como una categoría adicional.


```python
X = [['hombre', 'Safari'], ['mujer', None], [np.nan, 'Firefox']]
enc = OneHotEncoder(handle_unknown='error').fit(X)
enc.categories_
```

```python
enc.transform(X).toarray()
```

El parámetro `handle_unknown='error'` en la función `OneHotEncoder` define el comportamiento del codificador cuando se encuentra con categorías desconocidas durante la transformación de datos.

Específicamente:

`OneHotEncoder(handle_unknown='error')` configura el codificador para que genere un error cuando se encuentre con una categoría que no estaba presente en los datos de entrenamiento. **Usa el ejercicio anterior y compruébalo tu mismo!!**
